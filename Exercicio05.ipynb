1 Linha: Crie a classe Linha que tem dois atributos, coordenada1 e coordenada2.
Cada coordenada é uma tupla que carrega duas coordenadas cartesianas (x,y) que
denotam pontos do segmento de reta. Faça métodos que calculem o comprimento
do segmento de reta e sua inclinação.

```python
import math

class Linha:
    def __init__(self, coordenada1, coordenada2):
        self.coordenada1 = coordenada1
        self.coordenada2 = coordenada2
        
    def calcular_comprimento(self):
        x = (self.coordenada2[0] - self.coordenada1[0]) ** 2
        y = (self.coordenada2[1] - self.coordenada1[1]) ** 2
        d = math.sqrt(x + y)
        return d
    
    def calcular_inclinacao(self):
        x = (self.coordenada2[0] - self.coordenada1[0])
        y = (self.coordenada2[1] - self.coordenada1[1])
        i = y/x
        return i
```

2 Figuras: Crie a seguinte hierarquia de classes de figuras geométricas. Veja na figura
as fórmulas:
a. A classe abstrata Figura deve ter o método abstrato area.
b. A classe concreta Circulo é subclasse de Figura.
c. A classe abstrata Poligono é subclasse de Figura e deve ter os atributos base e
altura .
d. As classes concretas Triangulo, Losango, Retangulo e Quadrado são subclasses
de Poligono. Tente criar mais uma generalização aqui olhando as fórmulas da área.
e. Os polígonos Retangulo e Quadrado devem implementar a interface Diagonal,
que deve ter um método que calcula a diagonal.
f. Crie uma classe Geometria com uma lista de Figuras com pelo menos uma
figura de cada e imprima suas áreas, perímetros e diagonais.

```python
from abc import ABC, abstractmethod
import math

class Figura(ABC):
    @abstractmethod
    def calcula_area(self):
        pass
    
class Diagonal:
    @abstractmethod
    def calcula_diagonal(self):
        pass
    
class Circulo(Figura):
    def __init__(self, circunferencia):
        self.circunferencia = circunferencia
        
    def calcula_raio(self):
        r = self.circunferencia / (2 * math.pi)
        return r
    
    def calcula_area(self):
        a = self.calcula_raio() ** 2 * math.pi
        return a
        
class Poligono(Figura):
    
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
          
    def calcula_area(self):
        Figura.calcula_area(self)

class Triangulo(Poligono):
    
    def __init__(self, base, altura, lado1, lado2):
        Poligono.__init__(self, base, altura)
        self.lado1 = lado1
        self.lado2 = lado2
        
    def calcula_area(self):
        a = (self.base * self.altura)/2 
        return a
        
    def calcula_perimetro(self):
        p = self.base + self.lado1 + self.lado2 
        return p
        
class Losango(Poligono):
    
    def __init__(self, base, altura):
        Poligono.__init__(self, base, altura)
        
    def calcula_perimetro(self):
        p = self.base * 2 + self.altura * 2
        return p
        
class Retangulo(Poligono, Diagonal):
    
    def __init__(self, base, altura):
        Poligono.__init__(self, base, altura)
        
    def calcula_diagonal(self):
        x = self.base ** 2 + self.altura ** 2
        d = math.sqrt(x)
        return d
    
    def calcula_perimetro(self):
        p = self.base * 2 + self.altura * 2
        return p
        
    def calcula_area(self):
        a = self.base * self.altura 
        return a
        
class Quadrado(Poligono, Diagonal):
    
    def __init__(self, base, altura):
        Poligono.__init__(self, base, altura)
        
    def calcula_diagonal(self):
        x = self.base ** 2 + self.altura ** 2
        d = math.sqrt(x)
        return d
    
    def calcula_perimetro(self):
        p = self.base * 4
        return p
    
    def calcula_area(self):
        a = self.base * self.altura 
        return a

class Geometria:
    
    c = Circulo(15)
    t = Triangulo(10, 5, 6, 5)
    l = Losango(10, 10)
    r = Retangulo(10, 20)
    q = Quadrado(20, 20)
    
    print("Ciruclo raio",c.calcula_raio())
    print("Ciruclo area",c.calcula_area())
    
    print("Triangulo area",t.calcula_area())
    print("Triangulo perimetro",t.calcula_perimetro())
    
    print("Losango perimetro",l.calcula_perimetro())
    
    print("Retangulo diagonal",r.calcula_diagonal())
    print("Retangulo perimetro",r.calcula_perimetro())
    print("Retangulo area",r.calcula_area())
    
    print("Quadrado diagonal",q.calcula_diagonal())
    print("Quadrado perimetro",q.calcula_perimetro())
    print("Quadrado area",q.calcula_area())
```
